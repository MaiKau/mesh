```python
import pandas as pd
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pandas as pd
from numpy import nan

with open('meah.html', 'r') as f:
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')

href_list=[]

t1 = soup.find_all('a')
for t2 in t1:
    t3 = t2.get('href')
    t3 = 'https://meshb.nlm.nih.gov'+str(t3)
    href_list.append(t3)
    
del href_list[0]
href_list
```

```python
import pandas as pd
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pandas as pd
from numpy import nan
import time

df_combine = pd.DataFrame()


for i in range(len(href_list)):
    try:
        url = href_list[i]
#         time.sleep(2)
        res = requests.get(url)
#         time.sleep(2)
        soup = BeautifulSoup(res.text)

        soup = soup.find('div',{'class':'tab-pane active'})

        list_ = [child.text for child in soup.find_next("dl").children if isinstance(child, Tag)]
        list_ = [i.strip() for i in list_]

        MeSH_Heading = list_.index('MeSH Heading')
        Tree_Numbers = list_.index('Tree Number(s)')
        Unique_ID = list_.index('Unique ID')
        RDF_Unique_Identifier = list_.index('RDF Unique Identifier')
        Entry_Term = list_.index('Entry Term(s)')
        Date_Established = list_.index('Date Established')
        
        one = list_[MeSH_Heading:Tree_Numbers]
        two = list_[Tree_Numbers:Unique_ID]
        three = list_[Unique_ID:RDF_Unique_Identifier]
        four = list_[Entry_Term:Date_Established]

        df = pd.concat([pd.DataFrame(one),pd.DataFrame(three)],axis=1)
        df = pd.concat([df,pd.DataFrame(two)],axis=1)
        df = pd.concat([df,pd.DataFrame(four)],axis=1)
        df = df.reset_index()

        def change_df(df):
            arr = df.values
            new_df = pd.DataFrame(arr[1:,1:], index = arr[1:,0], columns = arr[0,1:])
            return(new_df)

        df = change_df(df)
        df['flag'] = str(i)

        try:
            df_combine = pd.concat([df_combine, df], axis=0)
            print('成功,'+str(i))
            if i % 5000 == 0:
                df_combine.to_excel(str(i)+'.xlsx')

        except:
            print('失敗,'+str(i))
            pass
        
    except:
        print('失敗,'+str(i))
        pass

df_combine.to_excel('df_combine_my.xlsx')
```
