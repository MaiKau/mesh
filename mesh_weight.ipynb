```python
import pandas as pd
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pandas as pd
from numpy import nan

with open('meah.html', 'r') as f:
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')

href_list=[]

t1 = soup.find_all('a')
for t2 in t1:
    t3 = t2.get('href')
    t3 = 'https://meshb.nlm.nih.gov'+str(t3)
    href_list.append(t3)
    
del href_list[0]
href_list
```

```python
len(href_list)
```

```python
df_combine = pd.DataFrame(columns=['Title','URL','QF_NAME','QF_URL'])

for i in range(len(href_list)):
    try:
        url = href_list[i]
        res = requests.get(url)
        soup = BeautifulSoup(res.text)
        title = soup.find('h1').get_text()
        title_remove = soup.find('h1').find_all('small')[0].extract().get_text()
        title = title.replace(title_remove,'').strip()

        soup = soup.find('span',{'id':'allowableQualifiers'})
        QFURL_list=[]
        text_list=[]

        a1 = soup.find_all('a')
        for k1 in a1:
            k2 = k1.get('href')
            k2 = 'https://meshb.nlm.nih.gov/record/ui?ui='+str(k2)
            QFURL_list.append(k2)

        for t1 in a1:
            t2 = t1.string.replace(' ', '').replace('\r', '').replace('\n', '')
            text_list.append(t2)    

        insert_df = pd.concat([pd.DataFrame(QFURL_list,columns=['QF_URL']),pd.DataFrame(text_list,columns=['QF_NAME'])],axis=1)
        insert_df['URL'] = url
        insert_df['Title'] = title
        df_combine = pd.concat([df_combine,insert_df],axis=0)
        print('成功',str(i))
        if i % 3000 == 0:
            df_combine.to_excel('weight_'+str(i)+'.xlsx')
        else:
            pass
    except:
        print('失敗',str(i))
        pass

df_combine.to_excel('df_combine_my.xlsx')
```
